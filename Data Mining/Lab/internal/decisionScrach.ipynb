```python
from math import log2

# 1-∑pi^2.
def gini_impurity(y):
    if(len(y)==0):
       return 0.0
    counts = np.unique(y,return_counts=True)
    probs = counts / counts.sum()
    return 1.0-np.sum(probs**2)

# p.logp
def entropy(y):
    if(len(y)==0): 
        return 0.0
    counts = np.unique(y,return_counts=True)
    probs = counts / counts.sum()
    return -np.sum(probs*np.log2(probs))

# 
def infromation_gain(y,y_left,y_right,criterion="gini"):
    if(criterion=="gini"): impurity = gini_impurity
    else: impurity=entropy

    H_parent = impurity(y)
    n=len(y)
    n_left=len(y_left)
    n_right=len(y_right)
    if(n_left==0 | n_right==0): return 0.0

    H_children = (n_left/n)*impurity(y_left) + (n_right/n) * impurity(y_right)
    return H_parent-H_children
```
