- Summary
    - The code loads the Iris dataset.
    - It trains a Decision Tree classifier using the data.
    - It uses the trained model to classify a new flower.
    - Decision Trees learn by repeatedly splitting features to separate classes.

```python
from sklearn import tree 
from sklearn.datasets import load_iris

dataset = load_iris()
X=dataset.data
Y=dataset.target

model = tree.DecisionTreeClassifier()
model.fit(X,Y)

print("Prediction : " , model.predict([[1.2 , 3 , 4 , 5.1]]))
```
